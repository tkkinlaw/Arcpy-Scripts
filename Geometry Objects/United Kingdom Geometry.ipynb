# Explore a complex geometry object - the United Kingdom

```python
fc = r"File path to a dataset of world territories"
```

If you use the Shape@ token, you will access the entire geometry object. If you used the Shape@XY token, you would retrieve the X and Y centroids of the polygon.

```python
fld = "Shape@"
exp = "NAME = 'United Kingdom'"
cursor = arcpy.da.SearchCursor(fc, fld, where_clause = exp)
for row in cursor:
    print(row)
```

This search cursor object contains a single feature, since we used the where_clause to extract only the UK. So, by calling upon the row variable we can access that single feature outside the for loop.

```python
row[0]
```

The UK is has multi part geometry. One feature made of many polygons.

```python
numPart = row[0].partCount
print("Specifically, this dataset records 27 polygons for the UK. (There are many many many more in real life. This is a simplified dataset.)")
```

If you use an index position on the polygon geometry object, you can access a single polygon of the UK.

```python
row[0][0]
```

Notice that the above code returns an array, containing all the point objects constituting that island. This is the main island of Great Britain.

You can use this array to create a polygon object of just the island of Great Britain.

```python
arcpy.Polygon(row[0][0])
```

You can continue to access this geometry object's properties and methods.

```python
arcpy.Polygon(row[0][0]).buffer(50000)
```
