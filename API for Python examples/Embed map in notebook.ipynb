# Embedding a map with ArcGIS API for Python

```python
import arcgis
from arcgis.gis import GIS
from getpass import getpass
```

```python
username = getpass()
password = getpass()
timsAGOLOrg = GIS("https://www.arcgis.com", username, password)
```

```python
dcBoundaryFL = timsAGOLOrg.content.get('dfc75cf4d1054bb580406052d4c66c7e')
```

```python
fl = dcBoundaryFL.layers[0]
queryResults = fl.query()
```

```python
queryResults.sdf
```

```python
map1 = timsAGOLOrg.map('DC')
map1.content.add(fl)
map1
```
