## Item graph basics

```python
from arcgis.gis import GIS
from arcgis.apps.itemgraph import create_dependency_graph
```

#### Connect to GIS

```python
gis = GIS(profile='PersonalOrg')
gis.users.me.username
```

#### Create dependency graph from an item

```python
id= '6fcc9dd39c2149ed9636dc8087c35ceb'
item = gis.content.get(id)
graph = create_dependency_graph(gis, [item])

print(f"There are {len(graph.all_items())} items in this graph")
```

#### List all items used by the target item/node

```python
targetNode = graph.get_node(id)
targetNode.requires()
```

#### Basic vizualization of the item graph

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
pos = nx.spring_layout(graph) 
nx.draw(graph, pos, with_labels=True, node_size=500, node_color="skyblue", font_size=6, font_color="green")
nx.draw_networkx_labels(graph, pos, labels=nx.get_node_attributes(graph, 'label'))
```
