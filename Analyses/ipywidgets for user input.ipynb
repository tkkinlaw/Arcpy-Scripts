```python
import ipywidgets as widgets
```

#### Get user input

This widget let's the end user provide the numeric value by draging the slider. This cell does not need run after dragging the slider.

```python
intWidget = widgets.IntSlider(
    value=7,
    min=0,
    max=10,
    step=1,
    description='Test:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
intWidget
```

```python
buffDist = f"{intWidget.value} Miles"
print(f"Buffering with {buffDist}")
arcpy.analysis.Buffer("GasStations", "BufferOutput", buffDist)
```
