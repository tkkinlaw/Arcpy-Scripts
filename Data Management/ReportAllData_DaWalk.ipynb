# Print information about all data in a workspace

```python
import arcpy
import os

workspace = r"C:\Users\tim10393\OneDrive - Esri\Courses\PYTS-8.0"
feature_classes = []

walk = arcpy.da.Walk(workspace, datatype="FeatureClass", type="Polygon")
```

```python
for dirpath, dirnames, filenames in walk:
    for filename in filenames:
        feature_classes.append(os.path.join(dirpath, filename))
```

```python
for fc in feature_classes:
    d = arcpy.Describe(fc)
    print(d.name)
    print(d.spatialReference.name)
    fieldsList = []
    for field in d.fields:
        fieldsList.append(field.name)
    print(f"Field Names: {fieldsList}")
    print('\n')
```
