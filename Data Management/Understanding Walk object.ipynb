```python
walk = arcpy.da.Walk()
for dirpath,dirnames,filenames in walk:
    print(f"Looking at: {dirpath}")
    if len(dirnames)==0:
        print("No other workspaces here.")
    elif len(dirnames)>0:
        print("Workspaces here:")
        for dirname in dirnames:
            print(f"- {dirname}")  
    if len(filenames)==0:
        print("There are no nonworkspaces here.")
    elif len(filenames)>0:
        print("Nonworkspaces here:")
        for filename in filenames:
            print(f"- {filename}") 
    print()
```
