# Seach for related values in Relationship Class

This script uses arcpy.da.SearchRelatedRecords, available starting in ArcGIS Pro 3.2, to search for the lots related to a record.

```python
import arcpy
import os
```

```python
relc = r"C:\Users\tim10393\Downloads\PADA_2.0_Oct20_Demos\DemoData\PADA\ParcelFabricsInApps\Polk_Initial_Lesson_8_Solution.gdb\ParcelEntry\RecordHasLot"
oids = [2]
originRecordsFields = ["OID@", "Name", "ParcelCount"]
destLotFields = ["OID@", "Name", "Shape_Area"]
```

```python
related_records = arcpy.da.SearchRelatedRecords(relc, oids, origin_fields=originRecordsFields, destination_fields=destLotFields)
for relatedRecords in related_records:
    record = relatedRecords[0]
    lot = relatedRecords[1]
    print(f"Record information:\n- OID: {record[0]}\n- Name: {record[1]},\n- Parcel count: {record[2]}")
    print(f"Lot information:\n- OID: {lot[0]}\n- Name: {lot[1]},\n- Area: {lot[2]}")
    print('\n')
del related_records
```
