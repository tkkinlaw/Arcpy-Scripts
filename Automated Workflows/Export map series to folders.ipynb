## Export map series pages to individual folders

This script assumes you are working with a map series with page numbers

```python
import arcpy
import os
```

```python
aprxPath = r"C:\DemoData\SNAP\Demos.aprx"
layoutName = "Ohio rivers map series"
outputRootFolder = r"C:\DemoData\SNAP\Layouts"
groupFieldName = 'SUBBASIN'
fileNameFieldName  = ''
resolution = 300
```

```python
os.makedirs(outputRootFolder, exist_ok=True)
```

```python
aprx = arcpy.mp.ArcGISProject(aprxPath)
layout = aprx.listLayouts(layoutName)[0]
mapSeries = layout.mapSeries
mapSeries.refresh()
```

```python
pageNameField = mapSeries.pageNameField.name
nameFieldToUse = fileNameFieldName if fileNameFieldName else pageNameField
```

```python
for pageNumber in range(1, mapSeries.pageCount + 1):
    mapSeries.currentPageNumber = pageNumber
    row = mapSeries.pageRow
    groupName = str(getattr(row, groupFieldName))
    pageName = str(getattr(row, nameFieldToUse))

    groupFolderPath = os.path.join(outputRootFolder, groupName)
    os.makedirs(groupFolderPath, exist_ok=True)
    outputPdfPath = os.path.join(groupFolderPath, f"{pageName}.pdf")

    layout.exportToPDF(out_pdf=outputPdfPath, resolution=resolution, image_quality="BEST", compress_vector_graphics=True,
                       image_compression="ADAPTIVE", embed_fonts=True, georef_info=True)
```
